## Compare Yield to Quoted Rate

We calculate the actual rate quoted at the moment of taking out the loan and compare it to the yield that was eventually earned on the loan.


```python
# We use the previous loan with ID 2604725

df_2604725 = df_payments[df_payments['SK_ID_PREV'] == 2604725].copy()
```

```python
df_2604725
```

```python
# What was the quoted rate on this loan?

details_2604725 = df_previous[df_previous['SK_ID_PREV'] == 2604725]
annuity = details_2604725['AMT_ANNUITY'].iloc[0]
credit = details_2604725['AMT_CREDIT'].iloc[0]
no_of_payments = details_2604725['CNT_PAYMENT'].iloc[0]
print("Annuity: %.2f" % annuity)
print("Credit: %.2f" % credit)
print("Number of monthly instalments: %i" % no_of_payments)
PMT = [-credit]
PMT = PMT + ([annuity] * int(no_of_payments))
irr = np.irr(PMT)
ear = (1+irr)**12 - 1
print("Annualized Rate on the loan: %.2f%%" % (100*ear))

```

```python
start_date = int(df_2604725['DAYS_INSTALMENT'].min() - 30)
print("Starting day: %i" % start_date)
end_date = int(df_2604725['DAYS_ENTRY_PAYMENT'].max())
print("Last payment on: %i" % end_date)
period_length = end_date - start_date
cf = [0] * (period_length + 1)
cf[0] = -credit
for i, row in df_2604725.iterrows():
    index = int(row['DAYS_ENTRY_PAYMENT'] - start_date)
    amount = row['AMT_INSTALMENT']
    cf[index] = amount
print(cf)
irr = np.irr(cf)
ear = (1+irr)**360 - 1 # we take 1 year to be 360 days
print("Annualized Rate on the loan: %.2f%%" % (100*ear))
```

So from above calculations we conclude that for loan with ID 2604725 the rate at which the loan was closed was 28.15% and in the end the yield was 28.72%
