Given the above specifications, what is the expected yield on this loan?

## Calculate the EAR of the loan upon application

```python
def ear(row):
    annuity = row['AMT_ANNUITY']
    credit = row['AMT_CREDIT']
    no_of_payments = row['CNT_PAYMENT']
    if np.isnan(annuity) or np.isnan(credit) or np.isnan(no_of_payments):
        return np.nan
    PMT = [-credit]
    PMT = PMT + ([annuity] * int(no_of_payments))
    irr = np.irr(PMT)
    ear = (1+irr)**12 - 1
    return ear

df_previous['EAR_OF_LOAN'] = df_previous.apply(lambda row:ear(row), axis = 1)
```
